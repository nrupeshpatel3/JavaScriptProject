```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


x = ["Public", "Private, Nonprofit",  "Proprietary", "Foreign Public", "Foreign Private",  "Foreign For-Profit"]
stype = [12.4, 6.4, 12.4, 4.0, 6.4, 6.0]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, stype, color='red')
plt.xlabel("School Type")

plt.ylabel("Default Rate")
plt.title("Student Loan Default Rates by School Type")
plt.ylim(0, 30) 

plt.xticks(x_pos, x)
plt.xticks(rotation=90)
plt.savefig("Images/School Type.png", bbox_inches = "tight")
plt.show()
```
