```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


x = ["Native American", "HBCU", "Hispanic", "Ethnicity Not Reported"]
ethnicity = [24.0, 16.4, 12.6, 9.9]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, ethnicity, color='red')
plt.xlabel("Ethnicity")

plt.ylabel("Default Rate")
plt.title("Student Loan Default Rates by Ethnicity")
plt.ylim(0, 30)

plt.xticks(x_pos, x)
plt.xticks(rotation=90)
plt.savefig("Images/Ethnicity.png", bbox_inches = "tight")
plt.show()
```
