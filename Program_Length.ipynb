```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


x = ["Grad/Prof≥ 300 hrs", "Non-Degree 600–899 hrs", "Non-Degree 900–1799 hrs", "Non-Degree 1800–2699 hrs", "Non-Degree ≥ 2700 hrs", "Associate's", "Bachelor's", "1st Prof Degree", "Master's or Doctor's",  "Two-Year Transfer"]
plength = [2.9, 13.3, 12.6, 10.2, 7.8, 15.4, 11, 2.4, 6, 4.6]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, plength, color='red')
plt.xlabel("Program Length")

plt.ylabel("Default Rate")
plt.title("Student Loan Default Rates by Program Length")
plt.ylim(0, 30)

plt.xticks(x_pos, x)
plt.xticks(rotation=90)

    
plt.savefig("Images/Program Length.png", bbox_inches = "tight")
plt.show()
```
